```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
files = {
    '2_5':  2.5,
    '5':    5.0,
    '7_5':  7.5,
    '10':   10.0,
    '12_5': 12.5,
    '15':   15.0,
    '20':   20.0,
    '25':   25.0,
    '30':   30.0,
    '35':   35.0,
    '40':   40.0,
    '45':   45.0,
    '50':   50.0,
    '55':   55.0}

mapping = dict()
err = dict()
data = dict()
for file_end, dist in files.items():
    f = pd.read_csv('sensor_raw_right_line_{0}.log'.format(file_end))
    data[dist] = f.right.values
    mapping[dist] = np.mean(f.right)
    err[dist] = 100*np.std(f.right)/np.mean(f.right)

fig, axs = plt.subplots(2, 1, figsize=(8, 8))

axs[0].plot(mapping.keys(), mapping.values(), '-o')
axs[0].set_xlabel('Distance [cm]'); axs[0].set_ylabel('mean')
axs[0].grid(True)

axs[1].plot(err.keys(), err.values(), '-o')
axs[1].set_xlabel('Distance [cm]'); axs[1].set_ylabel('std / mean [%]')
axs[1].grid(True)

plt.show()
```

```python
dist = 20.0
fig, axs = plt.subplots(figsize=(8, 5))
axs.hist(data[dist], 20); axs.grid(True)
axs.set_xlabel('raw_value at distance {0} cm'.format(dist));
```

```python
f = pd.read_csv('sensor_raw_right_line_sweep_2_5_to_55.log')
dt = 0.02
y = f.right
time = [t*dt for t in range(0, len(y))]

fig, axs = plt.subplots(figsize=(8, 5))
axs.plot(time, y, '-'); axs.grid(True)
axs.set_xlabel('time [s]'); axs.set_ylabel('raw_value');
```
